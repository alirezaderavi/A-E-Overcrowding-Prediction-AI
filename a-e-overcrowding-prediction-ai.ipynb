```python
path = "/kaggle/input/datasets/alirezaderavi98/a-and-e-overcrowding-monthly-nhs/"   

all_files = glob.glob(path + "*.csv")

dfs = []

def parse_period(p):
    if pd.isna(p):
        return pd.NaT
    
    try:
        return pd.to_datetime(p, format="%d/%m/%Y")
    except:
        pass
    
    match = re.search(r"MSitAE-([A-Z]+)-(\d{4})", str(p))
    if match:
        month = match.group(1).title()   # DECEMBER → December
        year = match.group(2)
        return pd.to_datetime(f"{month} {year}")
    
    return pd.NaT


for file in all_files:
    try:
        df = pd.read_csv(file)

        if "A&E attendances Type 1" in df.columns:

            df["Date"] = df["Period"].apply(parse_period)

            dfs.append(df)

    except Exception as e:
        print(f"Error reading {file}: {e}")


final_df = pd.concat(dfs, ignore_index=True)

final_df = final_df.sort_values("Date").reset_index(drop=True)

final_df.to_csv("A&E_Monthly_Master.csv", index=False)

print("Done! File saved as A&E_Monthly_Master.csv")
```

```python
def clean_ae_monthly(df):
    """
    Clean NHS A&E Monthly dataset
    """

    # ---------------------------------------
    # 1) Remove TOTAL rows (they break models)
    # ---------------------------------------
    df = df[df["Period"] != "TOTAL"]

    # ---------------------------------------
    # 2) Drop all 'Unnamed' columns (empty)
    # ---------------------------------------
    df = df.loc[:, ~df.columns.str.contains("Unnamed")]

    # ---------------------------------------
    # 3) Convert Period → real datetime
    #    Supports both:
    #    - MSitAE-DECEMBER-2024
    #    - 01/11/2024
    # ---------------------------------------
    def parse_period(p):
        if pd.isna(p):
            return pd.NaT
        
        # Case 1: Real date format (01/11/2024)
        try:
            return pd.to_datetime(p, format="%d/%m/%Y")
        except:
            pass
        
        # Case 2: MSitAE-MONTH-YEAR format
        match = re.search(r"MSitAE-([A-Z]+)-(\d{4})", str(p))
        if match:
            month = match.group(1).title()   # DECEMBER → December
            year = match.group(2)
            return pd.to_datetime(f"{month} {year}")
        
        return pd.NaT

    df["Date"] = df["Period"].apply(parse_period)

    # ---------------------------------------
    # 4) Remove rows with missing dates
    # ---------------------------------------
    df = df.dropna(subset=["Date"])

    # ---------------------------------------
    # 5) Convert numeric columns to numbers
    # ---------------------------------------
    numeric_cols = df.columns[df.columns.str.contains("A&E") |
                              df.columns.str.contains("admissions") |
                              df.columns.str.contains("attendances")]

    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors="coerce")

    # ---------------------------------------
    # 6) Remove fully empty rows
    # ---------------------------------------
    df = df.dropna(how="all")

    # ---------------------------------------
    # 7) Sort by date
    # ---------------------------------------
    df = df.sort_values("Date").reset_index(drop=True)

    return df
```

```python
master = clean_df 

master["Total_Attendances"] = (
    master["A&E attendances Type 1"] +
    master["A&E attendances Type 2"] +
    master["A&E attendances Other A&E Department"]
)

master["Total_Admissions"] = (
    master["Emergency admissions via A&E - Type 1"] +
    master["Emergency admissions via A&E - Type 2"] +
    master["Emergency admissions via A&E - Other A&E department"] +
    master["Other emergency admissions"]
)

master["Breach_Rate"] = (
    master["Attendances over 4hrs Type 1"] +
    master["Attendances over 4hrs Type 2"] +
    master["Attendances over 4hrs Other Department"]
) / master["Total_Attendances"]

master["DTA_Delay_Rate"] = (
    master["Patients who have waited 4-12 hs from DTA to admission"] +
    master["Patients who have waited 12+ hrs from DTA to admission"]
) / master["Total_Admissions"]

agg = master.groupby(["Org Code", "Date"]).agg({
    "Total_Attendances": "sum",
    "Total_Admissions": "sum",
    "Breach_Rate": "mean",
    "DTA_Delay_Rate": "mean"
}).reset_index()
```

```python
def feature_engineering(df):
    # -----------------------------
    # 1) Time-based features
    # -----------------------------
    df["Month"] = df["Date"].dt.month
    df["Year"] = df["Date"].dt.year
    df["Quarter"] = df["Date"].dt.quarter
    df["Week"] = df["Date"].dt.isocalendar().week.astype(int)
    df["Day"] = df["Date"].dt.day
    df["DayOfWeek"] = df["Date"].dt.dayofweek
    df["IsWeekend"] = df["DayOfWeek"].isin([5, 6]).astype(int)

    # Season (NHS-style)
    def season(month):
        if month in [12, 1, 2]:
            return "Winter"
        elif month in [3, 4, 5]:
            return "Spring"
        elif month in [6, 7, 8]:
            return "Summer"
        else:
            return "Autumn"

    df["Season"] = df["Month"].apply(season)

    # -----------------------------
    # 2) Attendance & Admission Features
    # -----------------------------
    df["Total_Attendances"] = (
        df["A&E attendances Type 1"] +
        df["A&E attendances Type 2"] +
        df["A&E attendances Other A&E Department"]
    )

    df["Total_Admissions"] = (
        df["Emergency admissions via A&E - Type 1"] +
        df["Emergency admissions via A&E - Type 2"] +
        df["Emergency admissions via A&E - Other A&E department"] +
        df["Other emergency admissions"]
    )

    # -----------------------------
    # 3) Breach Rate (Over 4hrs)
    # -----------------------------
    df["Total_Over4hrs"] = (
        df["Attendances over 4hrs Type 1"] +
        df["Attendances over 4hrs Type 2"] +
        df["Attendances over 4hrs Other Department"]
    )

    df["Breach_Rate"] = df["Total_Over4hrs"] / df["Total_Attendances"].replace(0, np.nan)

    # -----------------------------
    # 4) DTA Delay Rate
    # -----------------------------
    df["Total_DTA_Delay"] = (
        df["Patients who have waited 4-12 hs from DTA to admission"] +
        df["Patients who have waited 12+ hrs from DTA to admission"]
    )

    df["DTA_Delay_Rate"] = df["Total_DTA_Delay"] / df["Total_Admissions"].replace(0, np.nan)

    # -----------------------------
    # 5) Rolling Features (Hospital-level)
    # -----------------------------
    df = df.sort_values(["Org Code", "Date"])

    df["Attendances_Rolling_3"] = df.groupby("Org Code")["Total_Attendances"].rolling(3).mean().reset_index(0, drop=True)
    df["Attendances_Rolling_6"] = df.groupby("Org Code")["Total_Attendances"].rolling(6).mean().reset_index(0, drop=True)

    df["Admissions_Rolling_3"] = df.groupby("Org Code")["Total_Admissions"].rolling(3).mean().reset_index(0, drop=True)
    df["Admissions_Rolling_6"] = df.groupby("Org Code")["Total_Admissions"].rolling(6).mean().reset_index(0, drop=True)

    # -----------------------------
    # 6) Lag Features (برای مدل‌های ML)
    # -----------------------------
    df["Attendances_Lag_1"] = df.groupby("Org Code")["Total_Attendances"].shift(1)
    df["Attendances_Lag_2"] = df.groupby("Org Code")["Total_Attendances"].shift(2)
    df["Attendances_Lag_3"] = df.groupby("Org Code")["Total_Attendances"].shift(3)

    df["Admissions_Lag_1"] = df.groupby("Org Code")["Total_Admissions"].shift(1)
    df["Admissions_Lag_2"] = df.groupby("Org Code")["Total_Admissions"].shift(2)
    df["Admissions_Lag_3"] = df.groupby("Org Code")["Total_Admissions"].shift(3)

    # -----------------------------
    # 7) NHS Operational Flags
    # -----------------------------
    df["High_Breach_Flag"] = (df["Breach_Rate"] > 0.15).astype(int)
    df["High_DTA_Flag"] = (df["DTA_Delay_Rate"] > 0.10).astype(int)

    # -----------------------------
    # 8) Final Cleanup
    # -----------------------------
    df = df.replace([np.inf, -np.inf], np.nan)

    return df
```

```python
from prophet import Prophet

def build_prophet_model(df, org_code):
    data = df[df["Org Code"] == org_code].copy()
    data = data.sort_values("Date")

    prophet_df = data[["Date", "Total_Attendances"]].rename(columns={
        "Date": "ds",
        "Total_Attendances": "y"
    })

    model = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=False,
        daily_seasonality=False
    )
    model.fit(prophet_df)

    future = model.make_future_dataframe(periods=6, freq="MS")  # 6 ماه آینده
    forecast = model.predict(future)

    return model, forecast
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

def build_overcrowding_model(df):
    data = df[df["Total_Attendances"] > 0].copy()

    target = "High_Breach_Flag"
    features = [
        "Month", "Year", "Quarter", "Season",
        "Total_Attendances", "Total_Admissions",
        "Attendances_Rolling_3", "Attendances_Rolling_6",
        "Admissions_Rolling_3", "Admissions_Rolling_6",
        "Attendances_Lag_1", "Attendances_Lag_2", "Attendances_Lag_3",
        "DTA_Delay_Rate"
    ]

    X = data[features]
    y = data[target]

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, shuffle=True, stratify=y, random_state=42
    )

    model = LGBMClassifier(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=-1
    )

    model.fit(X_train, y_train)
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    auc = roc_auc_score(y_test, y_pred_proba)

    print("ROC-AUC:", auc)
    return model
```

```python
def bed_demand_estimator(df):
    df = df.copy()
    df["Beds_Demand_Estimate"] = (
        df["Total_Admissions"] * 1.0  
    )

    return df
```

```python
def ae_agent_decision(row):
    decisions = []

    if row["Breach_Rate"] > 0.15:
        decisions.append("Increase streaming capacity / triage staff")

    if row["DTA_Delay_Rate"] > 0.10:
        decisions.append("Escalate inpatient bed management / discharge planning")

    if row.get("Beds_Demand_Estimate", 0) > 0.9 * row.get("G&A_beds_available", 999999):
        decisions.append("Activate surge plan / open escalation beds")

    if not decisions:
        decisions.append("Status: Stable")

    return "; ".join(decisions)
```
